```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
palette=sns.color_palette('Set2')
```

```python
# This function will process the dataframe. It will replace the NaN with 0 and other categorical values 1,2,3,4,... 
#The intention is to make missing value 0 and other values to have a number. 

# df: dataframe to process
# categorical_cols: columns if df that are categorical

# it will return the processed datframe and the labels with corresponding numbers. 

def categorical_data_labeling_with_missing_value(df, categorical_cols):
  labels = {}
  df = df.fillna(0)
  for col in df.columns:
    if col in categorical_cols:
      dic_col = {} 
      unq = df[col].unique()
      if 0 in unq:
        dic_col['NaN'] = 0
        unq = np.delete(unq, np.where(unq==0))
      #print(unq)
      for i in range(len(unq)):
        df[col] = df[col].replace(unq[i], i+1)
        dic_col[unq[i]] = i+1;
      labels[col] = dic_col;
  return df, labels
```
